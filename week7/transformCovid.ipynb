```python
import pandas as pd

covid = pd.read_csv('COVID_county_data.csv')
covid['county'] = covid['county'] +', ' +covid['state']
covid['year'] = pd.DatetimeIndex(covid['date']).year
covid['month'] = pd.DatetimeIndex(covid['date']).month
covid.drop(columns=['date'], inplace=True)
covid = covid.groupby(['county', 'year', 'month']).agg({'cases': 'sum', 'deaths': 'sum'})
# covid['county'] = covid['county'] +', ' +covid['state']
covid

```
