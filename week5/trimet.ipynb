```python
import pandas as pd

df = pd.read_csv('trimet_03411_2.csv')
tempDF = pd.DataFrame()
templis = []
for i in range(len(df)):
    row = df.loc[i]
    reapeat = row['OCCURRENCES']
    for t in range(reapeat):
        templis.append(row)

newdf = pd.DataFrame(templis)
newdf.to_csv('new_trimet_03411_2.csv')
newdf
# newdf.fillna(value='N', axis=)
```
