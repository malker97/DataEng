```python
import pandas as pd
import numpy as np


df = pd.read_csv('books.csv')
df1 = df
needDrop = ['Edition Statement', 'Corporate Author', 'Corporate Contributors', 'Former owner', 'Engraver', 'Issuance type', 'Shelfmarks']
for colname in needDrop:
    df1 = df1.drop(columns=colname)
df1.head(5)
```

```python
df2 = df1
regex = r'^(\d{4})'
extr = df['Date of Publication'].str.extract(regex, expand=False)
df2['Date of Publication'] = pd.to_numeric(extr)
df2
```

```python
import re
stateCityU = []
with open('uniplaces.txt') as file:
    for line in file:
        if '[edit]' in line:
            state = line
        else:
            p1 = re.compile(r'[(](.*?)[)]', re.S)
            university = re.findall(p1, line)
            stateCityU.append((state,line,university))
universityInfoDf = pd.DataFrame(stateCityU, columns=['State', 'City', 'University'])
def get_citystate(item):
    if ' (' in item:
        return item[:item.find(' (')]
    elif '[' in item:
        return item[:item.find('[')]
    else:
        return item
towns_df =  universityInfoDf.applymap(get_citystate)
towns_df.to_csv('StateCityUniv.csv')
towns_df
```
